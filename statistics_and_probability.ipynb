## Hypothesis Testing Process

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. The process involves several steps:

1. **Formulate Hypotheses**:
   - **Null Hypothesis (H₀)**: A statement that there is no effect, difference, or relationship. It is the hypothesis that is tested.
   - **Alternative Hypothesis (H₁ or Ha)**: A statement that contradicts the null hypothesis, suggesting there is an effect, difference, or relationship.

2. **Select Significance Level (alpha)**: The significance level (often 0.05) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It defines the threshold for evidence required to reject H₀.

3. **Choose the Appropriate Statistical Test**: Depending on the data type and research question, choose an appropriate statistical test (e.g., t-test, chi-squared test, ANOVA).

4. **Collect Data and Calculate the Test Statistic**: Gather sample data and compute the test statistic, which summarizes the information about the sample. The test statistic depends on the type of hypothesis test being used.

5. **Calculate the p-value**: The p-value represents the probability of observing data as extreme as the sample data, assuming the null hypothesis is true. A smaller p-value indicates stronger evidence against H₀.

6. **Make a Decision**:
   - **Reject H₀** if the p-value ≤ alpha, meaning there is enough evidence to support the alternative hypothesis.
   - **Fail to Reject H₀** if the p-value > alpha, meaning there isn’t enough evidence to support the alternative hypothesis.

7. **Draw a Conclusion**: Based on the results, conclude whether the sample data provides sufficient evidence to reject the null hypothesis, supporting the alternative hypothesis.

This process helps make informed decisions about population characteristics based on sample data, with a defined level of confidence.

##T-Distribution vs. Normal Distribution

##T-Distribution:
The **T-distribution** (Student’s T-distribution) is a probability distribution used in statistics, particularly for estimating population parameters when the sample size is small and/or the population standard deviation is unknown.

### Key Characteristics:
- **Bell-shaped and symmetric**, similar to the normal distribution.
- **Heavier tails** than the normal distribution, meaning it has more probability in the tails.
- As **sample size (degrees of freedom) increases**, the T-distribution approaches the normal distribution.

## Differences Between T-Distribution and Normal Distribution:

| Feature             | T-Distribution                         | Normal Distribution                |
|--------------------|--------------------------------|--------------------------------|
| Shape             | Bell-shaped, heavier tails     | Bell-shaped, thinner tails    |
| Used When        | Sample size is small (n < 30) or population standard deviation is unknown | Population standard deviation is known and/or sample size is large |
| Degrees of Freedom (df) | Varies (df = n-1 for single sample) | Not applicable |
| Tail Behavior     | More probability in tails (thicker) | Less probability in tails (thinner) |

##When to Use the T-Distribution
- When working with **small sample sizes (n < 30)**.
- When **population standard deviation is unknown**.
- In hypothesis testing (e.g., **t-tests**).

As the **sample size increases**, the T-distribution **becomes more like the normal distribution**, making them almost identical for large samples.

50 students and finds the mean height to be 65 inches with a standard deviation of 3 inches.
Construct a 95% confidence interval for the population mean height.

\

      confidence interval = 95% at 95% wkt Zscore is 1.96
      \
      CI= [65+-1.96*(3/sqrt(n))]
      \

      CI= [65-0.8315, 65+0.8315]

      CI = [64.168 ,65.831]

```python
# Confidence interval for given data
#  from Given data
sample_mean = 65  # Sample mean height
sample_std = 3    # Sample standard deviation
sample_size = 50  # Sample size
confidence_level = 0.95  # Confidence level

# Z-score for 95% confidence interval
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Standard error
standard_error = sample_std / np.sqrt(sample_size)

# Margin of error
margin_of_error = z_score * standard_error

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print result
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

The normal distribution (bell curve) is symmetrical. That means when we say "95% confidence," we actually mean the middle 95% of the data, leaving 2.5% on each side (since 100% - 95% = 5%, and 5%/2 = 2.5% in each tail).

However, the function stats.norm.ppf() works with areas from 0 to 1 (cumulative probability from the left).

We want to find the Z-score that cuts off the top 2.5% of the curve.
That means we need to look at the lower 97.5% of the data.
\
Why Divide by 2?
We take (1 + confidence_level) / 2 to get this lower percentage.

For 95% confidence:
(1+0.95)/2=1.95/2=0.975
\
So, we use 0.975 in stats.norm.ppf() to get the Z-score 1.96.

###**Inference:** So the average height of the students and 95% confidence interval ranges from (64.17 , 65.83)

Test the manufacturer's claim at a significance level of 0.05 using a right-tailed hypothesis test.

```python
#t=test for right tail
# Given data
mu = 1000  # Claimed population mean
x_bar = 980  # Sample mean
s = 50  # Sample standard deviation
n = 50  # Sample size
alpha = 0.05  # Significance level

# Calculate t-statistic
t_stat = (x_bar - mu) / (s / np.sqrt(n))

# Find the critical t-value for right-tailed test
df = n - 1  # Degrees of freedom
t_critical = stats.t.ppf(1 - alpha, df)

# P-value
p_value = 1 - stats.t.cdf(t_stat, df)

# Print results
print(f"t_stat : {t_stat}")
print(f"t_critical : {t_critical}")
print(f"p_value : {p_value}")


```

10. A quality control manager at a factory wants to ensure that the average weight of products
coming off the production line is 500 grams. She takes a random sample of 30 products and
finds the mean weight to be 495 grams with a standard deviation of 10 grams. Test the
manager's claim at a significance level of 0.01 using a left-tailed hypothesis test.

### FROM GIVEN DATA

mu = 500  # population mean of average weights
\
x_bar = 495  # Sample mean of average weights
\
s = 10  # Sample standard deviation
\
n = 30  # Sample size
\
alpha = 0.01  # Significance level
\
t_stat = (x_bar - mu) / (s / sqrt(n))
\
t_stat = (495-500) / (10/ sqrt(30))
\
t_stat = -2.738
\
df = n - 1  
\
df= 29
\
p_value for -2.738 is 0.0052
\
Here null hypthesis is population mean of average weight of products is 500
\
alternate hypothes is since left tail population mean of average weight of products < 500

```python
# t_test for left tail
# Given data
mu = 500  # Claimed population mean
x_bar = 495  # Sample mean
s = 10  # Sample standard deviation
n = 30  # Sample size
alpha = 0.01  # Significance level

# Compute t-statistic
t_stat = (x_bar - mu) / (s / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Find the critical t-value for left-tailed test
t_critical = stats.t.ppf(alpha, df)

# Compute the p-value (left-tailed test)
p_value = stats.t.cdf(t_stat, df)

# Print results
print(f"T-Statistic: {t_stat:.4f}")
print(f"Critical T-Value: {t_critical:.4f}")
print(f"P-Value: {p_value:.4f}")

```

### 🔖 Inference : Since the p_value 0.0052 is less than 0.05 we reject the null hypothesis ✅
### Critical values gives us the cutoff for rejection region which is -2.4620 and | t_stat | > t_critical so we reject null hypothesis in this case
